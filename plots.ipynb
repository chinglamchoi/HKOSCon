```julia
using Plots
```

```julia
plot([1, 2, 3, 4], [1, 4, 2, 3]) 
```

```julia
x = range(0, 2, length=100)
plot(x,x, title = "Simple Plot", label = "linear")
plot!(x,x.^2, label = "quadratic")
plot!(x,x.^3, label = "cubic")
```

```julia
t = range(0, 2, length=100)
plot(t,sin.(2*pi*t).+1, title = "voltage against time", xlabel = "time (s)", ylabel = "voltage (mV)", legend = false)
```
